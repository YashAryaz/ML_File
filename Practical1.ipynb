# WAP to represent a data set and its single variate/multi-variate plots.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset 
data = pd.read_csv('cell_samples.csv')
```

```python
data
```

```python
# Single variable plot (e.g., histogram)
sns.histplot(data['Class'])
plt.title('Histogram of a Single Variable')
plt.show()

# Multi-variable plot (e.g., scatterplot)
sns.pairplot(data)
plt.title('Pairplot of Multiple Variables')
plt.show()
```
